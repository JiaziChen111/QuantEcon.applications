Let's start with standard imports

```python
import numpy as np
import matplotlib.pyplot as plt
from lucastree import LucasTree, compute_lt_price
```

## Exercise 1

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.set_xlabel(r'$y$', fontsize=16)
ax.set_ylabel(r'price', fontsize=16)

for beta in (.95, 0.98):
    print("Computing at beta = {}".format(beta))
    tree = LucasTree(beta=beta)
    grid = tree.grid
    price_vals = compute_lt_price(tree)
    label = r'$\beta = {}$'.format(beta)
    ax.plot(grid, price_vals, lw=2, alpha=0.7, label=label)

ax.legend(loc='upper left')
ax.set_xlim(min(grid), max(grid))

plt.show()
```
